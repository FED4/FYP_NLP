```python
import requests
from bs4 import BeautifulSoup
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
from webdriver_manager.utils import ChromeType
import time
import pandas as pd
import pickle

def child_contains_navstr(children_tags):
  func = lambda child: isinstance(child.string, NavigableString)
  return map(func, children_tags)

def end_node_with_text(tag, incl_embed_navstr=True):
  if tag.name not in tag:
    return False

  # check if the children nodes have deeper tag or 
  # just NavigableString, <em><b><strong> etc....
  nav_str_res = list(child_contains_navstr(tag.findChildren()))
  if all(nav_str_res) and nav_str_res:
    # print(children)
    return True if incl_embed_navstr else False

  # check if there it is a empty text node
  if not tag.text.replace('\n', ''):
    return False

  # check if there are some other tag wrapping around 
  if len(tag(text=False)) > 0:
    return False
  
  # final node
  return True

class Crawller:
    '''
    Crawller is initialized with a urls list or a url string.
    desired list or dictionary s returned by calling member functions.

    '''
    
    def __init__(self, urls, url=None):
        self.urls = urls
        self.soups = []
        self.url = url
        self.soup = None
        
        if url==None:
            #multiple urls
            for url in urls:
                driver = webdriver.Chrome(ChromeDriverManager(chrome_type=ChromeType.CHROMIUM).install())
                driver.get(url)

                # wait for the page to load
                wait = WebDriverWait(driver, 10)
                time.sleep(10)
                # get the page source
                page_source = driver.page_source

                driver.close()

                # parse the HTML
                soup = BeautifulSoup(page_source, "html.parser")
                self.soups.append(soup)
        else:
            #single url
            driver = webdriver.Chrome(ChromeDriverManager(chrome_type=ChromeType.CHROMIUM).install())
            driver.get(url)

            # wait for the page to load
            wait = WebDriverWait(driver, 10)
            time.sleep(10)
            # get the page source
            page_source = driver.page_source

            driver.close()

            # parse the HTML
            soup = BeautifulSoup(page_source, "html.parser")
            self.soup = soup
            
            
    def getNodesByTags(self, css_class_name, css_tag_name):
        '''
         extract content with tag name under class name
        input classname and tag name of html, return list of nodes
        '''
        items = []
        for soup in self.soups:
            #print(soup.prettify())
            sc = soup.find_all(class_=css_class_name)
            for m in sc:
                for i in range(len(m.select(css_tag_name))):
                    b = m.select(css_tag_name)[i].get_text().strip()
                    items.append(b)
        return list(set(items))
    
    def getNodesByClassName(self, css_class_name):
        '''
         extract content under class name
        input classname return list of nodes
        '''
        sc = self.soup.find_all(class_=css_class_name)
        for m in sc:
            b = m.get_text().strip()
        return b
    
    def gethref(self, css_class_name, css_tag_name, keyword=None):
        '''
        extract url given class name and tag name
        return dictionary of hyperlink text as key, url as value
        '''
        items = []
        for soup in self.soups:
            #print(soup.prettify())
            sc = soup.find_all(class_=css_class_name)
            print(sc)
            for m in sc:
                for i in range(len(m.select(css_tag_name))):
                    href = m.select(css_tag_name)[i].get('href')
                    content = m.select(css_tag_name)[i].get_text().strip()
                    #items[content] = (href)
                    items.append(href)
        return items
    
    def getEndNodes(self, tags):
        '''
        input list of acceptable tag names of html, return list of all end nodes
        '''
        texts = []
        for soup in self.soups:
            for tag in tags:
                content = soup.find_all(tag, recursive=True)

                for x in content:
                    lines = x.get_text().splitlines()
                    for unit in lines:
                        texts.append(unit.strip())
        return texts
    
    def getTableAsDataframe(self):
        try:
            table = self.soup.find_all('table')
            df = pd.read_html(str(table))[0]
            return df
        except Exception as e:
            print(e)
    
    def getSearchResults(self, items):
        '''
        input list of keywords, return search results in terms of a dictionary
        '''
        driver = webdriver.Chrome(ChromeDriverManager(chrome_type=ChromeType.CHROMIUM).install())
        driver.get("https://www.google.com/search?q="+items[0]+"&sxsrf=AOaemvKc3OO5LZKgsevbW5iXYk3KZXn_qw:1642721024031&source=lnms&tbm=shop&sa=X&ved=2ahUKEwjhnujYvMH1AhVUeMAKHa9oAHwQ_AUoAnoECAEQBA&biw=682&bih=759&dpr=2")

        #     iframe_elem = driver.find_element_by_css_selector('iframe[title="TrustArc Cookie Consent Manager"')
        #     driver.switch_to.frame(iframe_elem)
        #     accept_btn = driver.find_element_by_css_selector('.z')
        #     accept_btn.click()

        # NOTE: actual name of the button is required so remember to change 
        # the name from "Accept" to the proper button name
        WebDriverWait(driver, 100).until(EC.element_to_be_clickable((By.XPATH, "//span[text()='I agree']"))).click()
        db = {}
        for item in items:
            driver.get("https://www.google.com/search?q="+item+"&sxsrf=AOaemvKc3OO5LZKgsevbW5iXYk3KZXn_qw:1642721024031&source=lnms&tbm=shop&sa=X&ved=2ahUKEwjhnujYvMH1AhVUeMAKHa9oAHwQ_AUoAnoECAEQBA&biw=682&bih=759&dpr=2")
            content = driver.page_source

            soup = BeautifulSoup(content, 'html.parser')
            page = requests.get("https://www.google.com/search?q="+items[0]+"&sxsrf=AOaemvKc3OO5LZKgsevbW5iXYk3KZXn_qw:1642721024031&source=lnms&tbm=shop&sa=X&ved=2ahUKEwjhnujYvMH1AhVUeMAKHa9oAHwQ_AUoAnoECAEQBA&biw=682&bih=759&dpr=2")
            page
            #soup = BeautifulSoup(page.content, 'html.parser')
            sc = soup.find_all(class_="sh-pr__product-results-grid")
            out = []
            for m in sc:
                for b in m.select("div h4"):
                    out.append(b.get_text().strip())
            db[item]=out
        return db
```

```python
from fuzzywuzzy import process
import re
from csv import writer

df_nutrition1 = pd.read_pickle("../data/df_nutrition1")
df_nutrition2 = pd.read_pickle("../data/df_nutrition2")
df_nutrition3 = pd.read_pickle("../data/df_nutrition3")
df_nutrition4 = pd.read_pickle("../data/df_nutrition4")
df_nutrition5 = pd.read_pickle("../data/df_nutrition5")
df_nutrition6 = pd.read_pickle("../data/df_nutrition6")

class tableComprehensor:
    '''
    table comprehensor create a dictionary from table by defined rules

    '''
    
    def __init__(self):
        '''
        self.out = {}
        self.out['Product Name'] = []
        self.out['Product Description'] = []
        self.out['Energy'] = []
        self.out['Saturates'] = []
        self.out['Carbohydrate'] = []
        self.out['Sugars'] = []
        self.out['Protein'] = []
        self.out['Salt'] = []
           '''
        column_names = ['Product Name','Product Description','Category','Energy','Saturates','Carbohydrate','Sugars','Salt']
        with open('../data/nutrition_table.csv','a', newline='') as f_object:
            writer_object = writer(f_object)
            writer_object.writerow(column_names)
            f_object.close()
         
       
        
        
    def get_nutrition(self, df,product_name, product_description, category):
        '''
        take in the table of nutrition, output a list of values [Energy, Fat, Saturations, Carbohydrate, Sugars, Proteins, Salt] per 100g

        '''
        columnsNamesArr = df.columns.values
        rowsNamesArr = df.iloc[:, 0].values
        i = 0
        rows = {}
        for row in rowsNamesArr:
            rows[row] = i
            i = i+1
        
        per100 = process.extract('per 100g', columnsNamesArr,limit = 1)[0][0]
        energies = process.extract('energy', rowsNamesArr,limit = 2)
        kcal = process.extract('kcal', rowsNamesArr,limit = 1)
        candidates = energies+ kcal
        cells = {}
        energy_candidates = []
        for row in candidates:
            name = row[0]
            energy_index = rows[name]
            cell = df[per100][energy_index]
            if not name == cell:
                str_energy = ' '.join((per100, name, cell)) 
                print(str_energy)
                cells[str_energy] = cell
                energy_candidates.append(str_energy)
        
        kcal_found = process.extract('per 100g Energy kcal', energy_candidates,limit = 1)
        kcal_found = cells[kcal_found[0][0]]
        num = re.findall(r'\d+\.?\d*', kcal_found)[0]
        out = [product_name, product_description, category, num]
        df = df.transpose()
        
        #Replacing Header with Top Row
        new_header = df.iloc[0] #grab the first row for the header
        df = df[1:] #take the data less the header row
        df.columns = new_header #set the header row as the df header
        target_columns = ['Saturates','Carbohydrate', 'Sugars', 'Salt']
        for target in target_columns:
            matched_nutrition = process.extract(target, rowsNamesArr,limit = 1)
            specific_nutrition = df[matched_nutrition[0][0]].transpose()
            value = specific_nutrition[per100]
            num = re.findall(r'\d+\.?\d*', value)[0]
            out.append(num)
        print(out)
        with open('../data/nutrition_table.csv','a', newline='') as f_object:
            writer_object = writer(f_object)
            writer_object.writerow(out)
            f_object.close()
         
#tc = tableComprehensor()
#tc.get_nutrition(df_nutrition5,'aa', 'bb')
```
